```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las señales de entrada
x = np.array([1, 2, 3, 4])
h = np.array([0.5, 0.5, 0.5])

# Realizar la convolución
y = np.convolve(x, h)

# Crear los ejes x para cada señal
x_axis_x = np.arange(0, len(x))
x_axis_h = np.arange(0, len(h))
x_axis_y = np.arange(0, len(y))

# Graficar las señales
plt.figure(figsize=(10, 10))

plt.subplot(3, 1, 1)
plt.stem(x_axis_x, x, use_line_collection=True)
plt.title('Señal de entrada x')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')

plt.subplot(3, 1, 2)
plt.stem(x_axis_h, h, use_line_collection=True)
plt.title('Señal de filtro h')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')

plt.subplot(3, 1, 3)
plt.stem(x_axis_y, y, use_line_collection=True)
plt.title('Señal de salida y')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```
