```python
import numpy as np
import matplotlib.pyplot as plt

# Definir seales de entrada
t = np.linspace(0, 2*np.pi, 1000) # Vector de tiempo
x1 = np.sin(4*t) + np.random.normal(0, 0.3, len(t)) # Seal 1
x2 = np.cos(2*t) + np.random.normal(0, 0.3, len(t)) # Seal 2

# Correlacin cruzada
corr = np.correlate(x1, x2, mode='same')

# Graficar seales de entrada y correlacin cruzada
plt.figure(figsize=(10,10))
plt.subplot(311)
plt.plot(t, x1, 'b', label='Seal 1')
plt.legend(loc='upper right')
plt.subplot(312)
plt.plot(t, x2, 'r', label='Seal 2')
plt.legend(loc='upper right')
plt.subplot(313)
plt.plot(t, corr, 'g', label='Correlacin cruzada')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

```
