El teorema de Nyquist-Shannon establece que para evitar la pérdida de información en la reconstrucción de una señal continua a partir de su versión muestreada, la frecuencia de muestreo debe ser al menos el doble de la frecuencia máxima presente en la señal original.

Matemáticamente, el teorema de Nyquist se puede expresar como:

fs > 2*fmax

Donde fs es la frecuencia de muestreo y f_max es la frecuencia máxima presente en la señal original.

Esto significa que si la señal original tiene componentes de frecuencia de hasta 10 kHz, por ejemplo, la frecuencia de muestreo debe ser de al menos 20 kHz para evitar la pérdida de información en el proceso de muestreo y reconstrucción.

El teorema de Nyquist es importante en la teoría de la señal y en la práctica del procesamiento de señales, ya que garantiza que la información de la señal original se pueda recuperar de manera exacta a partir de su versión muestreada, siempre que se cumpla la condición de muestreo adecuado.

Además, el teorema de Nyquist también establece que si la frecuencia de muestreo es menor que el doble de la frecuencia máxima, se producirá un fenómeno conocido como aliasing, donde las frecuencias más altas se muestrean como frecuencias más bajas, lo que puede causar distorsiones en la señal muestreada. Por lo tanto, es importante considerar la frecuencia de muestreo adecuada para evitar el aliasing y garantizar una buena calidad de la señal.

```python
import numpy as np
import matplotlib.pyplot as plt

fs1=20
f1=1
t1=np.arange(0,5,1.0/fs1)
x1=np.sin(2*np.pi*f1*t1)

fs2=2 #Según el teorema de Nyquist es el doble de la frecuencia f2
f2=1
t2=np.arange(0.25,10,1.0/fs2)
x2=np.sin(2*np.pi*f2*t2)

plt.figure(figsize=(10,4))
plt.plot(t1,x1,"b-.",label="Señal con fs=20Hz")
plt.plot(t2,x2,"r-*",label="Señal con fs=2Hz")
plt.xlabel("Tiempo[s]")
plt.ylabel("Amplitud")
plt.grid()
plt.legend(loc="best")
plt.title("Señales sinusoidales")
plt.xlim(min(t1),max(t1))
plt.show()
```

Como se puede ver en la imagen la senal roja cumple y demuestra que el teorema es cierto solo con el doble vemos que podemos tener una nocion de como es la senal original
