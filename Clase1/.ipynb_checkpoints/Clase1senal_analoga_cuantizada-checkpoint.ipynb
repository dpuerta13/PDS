```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una señal diente de sierra
fs = 1000 # Frecuencia de muestreo (Hz)
t = np.arange(0, 1, 1/fs) # Vector de tiempo
f = 5 # Frecuencia de la señal (Hz)
x = 2 * (t * f - np.floor(t * f + 0.5)) # Señal diente de sierra

# Cuantizar la señal a 4 bits
bits = 4 # se puede cambiar la cantidad de bits que se toman para la cuantizacion
quantized = np.round((x + 1) * ((2**bits)-1) / 2) / ((2**bits)-1) * 2 - 1

# Graficar las señales original y cuantizada
plt.figure(figsize=(16,4))
plt.plot(t, x, label='Original')
plt.plot(t, quantized, label='Cuantizada')
plt.legend()
plt.grid()
plt.show()
```
