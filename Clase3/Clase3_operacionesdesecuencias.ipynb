```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las secuencias de tiempo discreto
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Suma de las secuencias
suma = x + y

# Resta de las secuencias
resta = x - y

# Convoluci√≥n de las secuencias
convolucion = np.convolve(x, y)

# Graficamos las secuencias y los resultados
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.stem(x)
plt.title('x[n]')
plt.subplot(2, 2, 2)
plt.stem(y)
plt.title('y[n]')
plt.subplot(2, 2, 3)
plt.stem(suma)
plt.title('x[n] + y[n]')
plt.subplot(2, 2, 4)
plt.stem(convolucion)
plt.title('x[n] * y[n]')
plt.show()
```
